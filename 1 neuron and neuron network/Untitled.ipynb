Результат работы нейрона -- это функция активации взятая от суммы входов нейрона перемноженных с весами, плюс смещение. 

То есть это означает, что до тех пор, пока в нейроне не накопилось достаточное количество заряда, ничего не происходит, и когда там накопилось достаточное количество заряда нейрон выстреливает какой-то сигнал. 


У этой пороговой функции активации есть два сильно отличающихся друг от друга значения: 0 и 1.

Место, где функция активации изменяет своё значение с нуля на единицу, мы будем называть разделяющей поверхностью. Разделяющая поверхность находится там, где аргумент функции активации равен нулю. 
Разделяющая поверхность -- это какая-то плоскость, которая задаётся вектором w и смещением b.

В случае, когда у нейрона 2 входа, разделяющая поверхность в координатах $x_0, x_1$это прямая <br><br>
Разделяющая поверхность - это геометрическое место точек, где функция активации меняет свое значение (в любой ϵ окрестности любой из таких точек находятся два различных значения функции активации)<br><br>
Разделяющая поверхность задается формулой $\left< \overrightarrow{x}, \overrightarrow{w} \right> + b = 0⟨x, w⟩+b=0$ <br><br>
Разделяющая поверхность является подмножеством пространства векторов $\overrightarrow{x}$
